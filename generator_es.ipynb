```python
from pydub import AudioSegment
import numpy as np
```

```python
mywebm = AudioSegment.from_file('/content/drive/MyDrive/drive-download-20230327T031322Z-001/긱사3/sample-1.webm', format = 'webm')
```

```python
data = np.array(mywebm.get_array_of_samples())
print(data.shape)
print(np.max(data))
```

```python
data = data[::6] / np.max(np.abs(data[::6]))
print(data.shape)
print(np.max(data))

```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data)
plt.show()
```
